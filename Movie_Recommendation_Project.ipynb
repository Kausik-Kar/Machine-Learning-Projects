```python
new_movie = movie_dataset.iloc[0:100, :]
new_movie.shape
```

```python
movie_mat = new_movie.pivot_table(index = 'id', columns = 'original_title', values = 'vote_average')
movie_mat.head()
```

```python
white = movie_mat['White Squall']
white
```

```python
similar_white_squall = movie_mat.corrwith(white)
similar_white_squall
```

```python
similar_white_squall = pd.DataFrame(similar_white_squall)
similar_white_squall.head()
```
