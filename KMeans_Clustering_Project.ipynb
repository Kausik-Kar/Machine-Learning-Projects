```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
data
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey = True, figsize = (10,6))
ax1.set_title("KMeans")
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_)

ax2.set_title("Actual")
ax2.scatter(data[0][:,0],data[0][:,1], c = data[1])
```

```python
import pandas as pd

df = pd.read_csv('YESBANK.NS.csv')
df.head()
```

```python
input_data = df[['Open', 'Low']].values
input_data
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6,6))
plt.scatter(df['Open'], df['Low'])
plt.xlabel("Open")
plt.ylabel("Low")
plt.show()
```

```python
model = KMeans(n_clusters = 2)
model.fit(input_data)
```

```python
plt.figure(figsize = (6,6))
plt.scatter(df['Open'], df['Low'], c = model.labels_)
plt.xlabel("Open")
plt.ylabel("Low")
plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['Targets'] = iris.target
df.head()
```

```python
df.tail()
```

```python
input_data1 = df.iloc[:, [0,2]]
input_data1.head()
```

```python
groupings = df['Targets'].values
groupings
```

```python
from sklearn.cluster import KMeans
clusterModel = KMeans(n_clusters = 3)
clusterModel.fit(input_data1)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey = True, figsize = (10,10))

ax1.set_title("KMeans")
ax1.scatter(input_data1['sepal length (cm)'], input_data1['petal length (cm)'], c = clusterModel.labels_)

ax2.set_title("Actual")
ax2.scatter(input_data1['sepal length (cm)'], input_data1['petal length (cm)'], c = df['Targets'])
```
