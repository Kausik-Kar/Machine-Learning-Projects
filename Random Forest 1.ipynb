```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
#for digits dataset
digits = load_digits()
```

```python
digits.keys()
```

```python
plt.gray()

for i in range(4):
    plt.matshow(digits.images[i])
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
model = RandomForestClassifier(n_estimators = 100)

from sklearn.model_selection import train_test_split

z = df.drop(['target'], axis = 'columns')
```

```python
x_train, x_test, y_train, y_test = train_test_split(z, digits.target, test_size = 0.25)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns

plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted Values')
plt.ylabel('Truth Values')
```

```python
#Next is Iris Dataset
from sklearn.datasets import load_iris
iris = load_iris()

iris.keys()
```

```python
df = pd.DataFrame(iris.data)
df.head()
```

```python
df['Target'] = iris.target
df.head()
```

```python
xd = df.drop(['Target'], axis = 'columns')

x_train, x_test, y_train, y_test = train_test_split(xd, iris.target, test_size = 0.35)
```

```python
model1 = RandomForestClassifier(n_estimators = 10)
model1.fit(x_train, y_train)
```

```python
model1.score(x_test, y_test)
```

```python
y_pred = model1.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.figure(figsize = (5,4))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted Values')
plt.ylabel('Truth Values')
```
