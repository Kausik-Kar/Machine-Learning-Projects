```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['Target'] = iris.target
df.tail()
```

```python
rep = set(iris.target.tolist())
rep
```

```python
rep = list(rep)
rep
```

```python
names = iris.target_names.tolist()
names
```

```python
dt = list(zip(rep,names))
dt
```

```python
cat_names = []

for i in df['Target']:
  cat_names.append(dt[i][1])

cat_names
```

```python
x = df['sepal length (cm)'].tolist()
x
```

```python
y = df['petal width (cm)'].tolist()
y
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.35)
```

```python
W = tf.Variable(3.0)
b = tf.Variable(2.0)
```

```python
train_steps = 500
learn_rate = 0.00092
```

```python
for i in range(train_steps):
  with tf.GradientTape() as tape:

    output_step = x_train*W + b

    error = output_step - y_train
    loss = tf.reduce_mean(tf.square(error))

    dW, db = tape.gradient(loss, [W,b])

    W.assign_sub(dW*learn_rate)
    b.assign_sub(db*learn_rate)

  if i%20 == 0:
    print(f"At step {i} the loss is {loss}.")
```

```python
y_pred = x_test*W + b
y_pred
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (6,6))
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, c = 'red')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.show()
```
