```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

housing.keys()
```

```python
df = pd.DataFrame(housing.data, columns = housing.feature_names)
df.head()
```

```python
trg = pd.DataFrame(housing.target, columns = housing.target_names)
trg.head()
```

```python
df['MedHouseVal'] = trg['MedHouseVal']
df.head()
```

```python
df = df.drop('AveRooms', axis = 1)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[['MedInc', 'HouseAge', 'AveBedrms', 'Population']].values
Y = df['MedHouseVal'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.25)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
x_test
```

```python
plt.scatter(x_test[:, 0], y_pred, c = 'red')
plt.scatter(x_test[:, 0], y_test, c = 'blue')
plt.show()
```
